```python
# 첫번째 과제
import pandas as pd
import numpy as np

# github의 자동차생산자료를 읽어들인다
path = 'https://github.com/dongupak/DataML/raw/main/csv/'
file = path + 'vehicle_prod.csv'
df = pd.read_csv(file, index_col=0)

# 데이터프레임에서 국가명을 제외한 숫자의 합을 구하여 Total로 명명
sums = df.select_dtypes(np.number).sum().rename('Total')
sums_df = pd.DataFrame(sums).T
total_df = pd.concat([df, sums_df])

# Average 열 추가
averages = df.select_dtypes(np.number).mean().rename('Average')
averages_df = pd.DataFrame(averages).T
total_df = pd.concat([total_df, averages_df])

# 결과 출력
total_df
```

```python
# 두번째 과제
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 데이터 파일 읽기
weather_file = 'busan_weather_file.csv'  # 파일 경로를 지정하세요
weather = pd.read_csv(weather_file, encoding='CP949')

# '일시' 열로부터 월 정보를 추출
weather['month'] = weather['일시']  # '일시'가 월을 나타낸다고 가정

# 월별 평균풍속 계산
monthly_wind = []

for i in range(1, 13):
    monthly_data = weather[weather['month'] == i]
    if not monthly_data.empty:
        monthly_wind.append(monthly_data['평균풍속'].mean())
    else:
        monthly_wind.append(0)  # 데이터가 없는 경우 0으로 설정

months = np.arange(1, 13)

# 평균풍속 그래프 그리기
plt.bar(months, monthly_wind, color='blue')

# 그래프 레이블 및 제목 설정
plt.xlabel('Month')
plt.ylabel('Average Wind Speed')
plt.title('Monthly Average Wind Speed')
plt.xticks(months)  # X축 ticks 설정
plt.ylim(0, max(monthly_wind) + 1)  # Y축 범위 설정
plt.show()
```
